```python
from collections import defaultdict

team_recent_results = defaultdict(list)
```

```python
for idx, row in match_df.iterrows():
    
    team_a = row["Bat First"]
    team_b = row["Bat Second"]
    winner = row["match_won_by"]
    
    # Team A recent form
    recent_a = team_recent_results[team_a][-5:]
    
    if len(recent_a) > 0:
        match_df.loc[idx, "team_a_recent_win_pct"] = (
            (sum(recent_a) + 1) / (len(recent_a) + 2)
        )
    else:
        match_df.loc[idx, "team_a_recent_win_pct"] = 0.5
    
    # Team B recent form
    recent_b = team_recent_results[team_b][-5:]
    
    if len(recent_b) > 0:
        match_df.loc[idx, "team_b_recent_win_pct"] = (
            (sum(recent_b) + 1) / (len(recent_b) + 2)
        )
    else:
        match_df.loc[idx, "team_b_recent_win_pct"] = 0.5
    
    # Update AFTER match
    if winner == team_a:
        team_recent_results[team_a].append(1)
        team_recent_results[team_b].append(0)
    else:
        team_recent_results[team_a].append(0)
        team_recent_results[team_b].append(1)
```

```python
match_df["recent_form_diff"] = (
    match_df["team_a_recent_win_pct"]
    - match_df["team_b_recent_win_pct"]
)
```

```python
venue_history = {}
```

```python
for idx, row in match_df.iterrows():
    
    venue = row["Venue"]
    team_a = row["Bat First"]
    team_b = row["Bat Second"]
    winner = row["match_won_by"]
    
    # Create keys
    key_a = (team_a, venue)
    key_b = (team_b, venue)
    
    # Initialize if missing
    for key in [key_a, key_b]:
        if key not in venue_history:
            venue_history[key] = {
                "matches": 0,
                "wins": 0
            }
    
    # Team A venue win %
    matches_a = venue_history[key_a]["matches"]
    wins_a = venue_history[key_a]["wins"]
    
    match_df.loc[idx, "team_a_venue_win_pct"] = (
        (wins_a + 1)
        /
        (matches_a + 2)
    )
    
    # Team B venue win %
    matches_b = venue_history[key_b]["matches"]
    wins_b = venue_history[key_b]["wins"]
    
    match_df.loc[idx, "team_b_venue_win_pct"] = (
        (wins_b + 1)
        /
        (matches_b + 2)
    )
    
    # Update AFTER match
    venue_history[key_a]["matches"] += 1
    venue_history[key_b]["matches"] += 1
    
    if winner == team_a:
        venue_history[key_a]["wins"] += 1
    else:
        venue_history[key_b]["wins"] += 1
```

```python
for idx, row in match_df.iterrows():
    
    team_a = row["Team_A"]
    team_b = row["Team_B"]
    winner = row["match_won_by"]
    
    # Initialize teams
    for team in [team_a, team_b]:
        if team not in chase_history:
            chase_history[team] = {
                "chase_matches": 0,
                "chase_wins": 0
            }
    
    # Team A chase stats BEFORE current match
    matches_a = chase_history[team_a]["chase_matches"]
    wins_a = chase_history[team_a]["chase_wins"]
    
    match_df.loc[idx, "team_a_chase_win_pct"] = (
        (wins_a + 1)
        /
        (matches_a + 2)
    )
    
    # Team B chase stats BEFORE current match
    matches_b = chase_history[team_b]["chase_matches"]
    wins_b = chase_history[team_b]["chase_wins"]
    
    match_df.loc[idx, "team_b_chase_win_pct"] = (
        (wins_b + 1)
        /
        (matches_b + 2)
    )
    
    # UPDATE AFTER MATCH
    
    # Team batting second always chased
    chasing_team = row["Bat Second"]
    
    chase_history[chasing_team]["chase_matches"] += 1
    
    # If chasing team won
    if winner == chasing_team:
        chase_history[chasing_team]["chase_wins"] += 1
```

```python
for idx, row in match_df.iterrows():
    
    team_a = row["Team_A"]
    team_b = row["Team_B"]
    winner = row["match_won_by"]
    
    # Initialize teams
    for team in [team_a, team_b]:
        if team not in team_history:
            team_history[team] = {
                "matches": 0,
                "wins": 0
            }
    
    # Team A past win %
    matches_a = team_history[team_a]["matches"]
    wins_a = team_history[team_a]["wins"]
    
    match_df.loc[idx, "team_a_past_win_pct"] = (
        (wins_a + 1)
        /
        (matches_a + 2)
    )
    
    # Team B past win %
    matches_b = team_history[team_b]["matches"]
    wins_b = team_history[team_b]["wins"]
    
    match_df.loc[idx, "team_b_past_win_pct"] = (
        (wins_b + 1)
        /
        (matches_b + 2)
    )
    
    # UPDATE AFTER MATCH
    
    team_history[team_a]["matches"] += 1
    team_history[team_b]["matches"] += 1
    
    if winner == team_a:
        team_history[team_a]["wins"] += 1
    else:
        team_history[team_b]["wins"] += 1
```
